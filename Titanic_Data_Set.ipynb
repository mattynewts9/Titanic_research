```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset("titanic")
# Loading the Data 
```

```python
print("First 5 rows of the dataset:")
print(df.head())
```

```python
survival_count = plt.figure(figsize=(8, 5))
sns.countplot(x='survived', hue="survived", data=df, palette='Set2', legend='auto')
plt.title("Survival Count")
plt.show()
```

```python
gender_survival_differences = sns.barplot(x="sex", hue="sex", y="survived", data=df, palette='Set1', legend='auto') 
plt.title("Difference Between Genders")
```

```python
class_based_survival = sns.pointplot(data=df, x='class', y='survived', color='navy', markers='o', markersize=10, linestyle='-')
plt.title('How class affected survival rates')
```

```python
age_survival_comparison_data = sns.kdeplot(data=df, x='age', hue='survived', fill=True)
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.barplot(data=df, x="survived", y="age", hue="class")
plt.title("Indepth survival data by age & class")
plt.ylabel('Survival Rate')
plt.show()
```
